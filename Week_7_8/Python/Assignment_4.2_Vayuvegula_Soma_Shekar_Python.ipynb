```python
#Importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as py
```

```python
#Read csv file
df_crime=pd.read_csv("crimerates-by-state-2005.csv")
```

```python
df_crime.head()
```

# Scatter Plot

```python
plt.figure(figsize=(15,8))
ax=sns.scatterplot(data=df_crime,x="state",y="burglary").set(title="Burglary vs State - Scatter Plot")
locs,labels=plt.xticks()
plt.setp(labels,rotation=90)
plt.show()
```

# Bubble Plot

```python
plt.figure(figsize=(15,8))
ax=sns.scatterplot(data=df_crime,x="state",y="burglary", size="burglary",legend=False,sizes=(20,2000)).set(title="Burglary vs State - Bubble Plot")
locs,labels=plt.xticks()
plt.setp(labels,rotation=90)
plt.show()
```

# Density Plot

```python
plt.figure(figsize=(15,8))
sns.displot(x=df_crime["burglary"],kde=True).set(title="Burglary - Density Plot")
locs,labels=plt.xticks()
plt.setp(labels,rotation=90)
plt.show()
```
